# Building Recommendation System  (with NaN values)

```python
c=df['vote_average'].mean()
c
```

```python
m=df['vote_count'].quantile(0.9)
m
```

```python
qm=df[(df['runtime']>150)&(df['revenue']>300000000)&(df['homepage'].notna())]
qm=qm[qm['vote_count']>=m]
qm.shape
```

```python
qm.head()
```

```python
def wr(x,m=m,c=c):
    v=x['vote_count']
    r=x['vote_average']
    return (v/(v+m)*r)+(m/(m+v)*c)
```

```python
qm['score']=qm.apply(wr,axis=1)
qm=qm.sort_values('score',ascending=False)
qm[['title','vote_count','vote_average','score']].head(10)
```

```python
d=df.sort_values('vote_count',ascending=False)
plt.figure(figsize=(18,6))

plt.barh(d['title'].head(15),d['vote_count'].head(15),align='center',color='cornflowerblue')
plt.gca().invert_yaxis()
plt.xlabel("Votes")
plt.title("Most voted Movies")
```

# Building Recommendation System (by dropping NaN values)

```python
df.isna().sum()
```

```python
df1 = df.dropna()
df1.isna().sum()
```

```python
c1=df1['vote_average'].mean()
c1
```

```python
m1=df1['vote_count'].quantile(0.9)
m1
```

```python
qm1=df1[(df1['runtime']>150)&(df1['revenue']>300000000)]
qm1=qm1[qm1['vote_count']>=m1]
qm1.head()
```

```python
qm1['score']=qm1.apply(wr,axis=1)
qm1.head()
```

```python
qm1=qm1.sort_values('score',ascending=False)
qm1[['title','vote_count','vote_average','score']].head(10)
```
